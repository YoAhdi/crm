```python
import os
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
```

```python
accounts_path = 'Accounts List.csv'
leads_path = 'Leads List.csv'
accounts_df = pd.read_csv(accounts_path)
leads_df = pd.read_csv(leads_path)
```

```python
leads_df[leads_df['Company Name'].isna()]
```

```python
accounts_duplicates = accounts_df[accounts_df['Company Name'].duplicated()]['Company Name']
accounts_df[accounts_df['Company Name'].isin(accounts_duplicates)].sort_values(by='Company Name', ascending=True).reset_index(drop=True)
```

```python
leads_duplicates = leads_df[leads_df['Company Name'].duplicated()]['Company Name']
leads_df[leads_df['Company Name'].isin(accounts_duplicates)].sort_values(by='Company Name', ascending=True).reset_index(drop=True)
```

```python
a_i = accounts_df[accounts_df['Company Name'].isna()].index
l_i = leads_df[leads_df['Company Name'].isna()].index
accounts_df.loc[a_i, 'Company Name'] = 'Trek Bicycle'
leads_df.loc[l_i, 'Company Name'] = 'Trek Bicycle'
```

```python
accounts_df =  accounts_df.sort_values(by='Company Name').reset_index(drop=True)
accounts_df['AccountID'] = range(0, accounts_df.shape[0])
accounts_df['AccountID'] = 1000 + (accounts_df.index * 10)
accounts_df['Company Revenue'] = accounts_df['Company Revenue'].astype(int)
```

```python
accounts_df.rename(columns={' Empoyee Head Count in millions': 'Employee Head Count'}, inplace=True)
accounts_df.rename(columns={'Company Revenue': 'Company Revenue (in millions)'}, inplace=True)
accounts_df.drop_duplicates(subset='Company Name', keep='first', inplace=True)
accounts_df.reset_index(drop=True, inplace=True)
```

```python
leads_df = leads_df.sort_values(by='Company Name', ascending=True).reset_index(drop=True)
leads_df = pd.merge(leads_df, accounts_df[['Company Name', 'AccountID']], on='Company Name', how='left')
leads_df['AccountID'] = leads_df['AccountID'].astype('Int64')
leads_df['Company Revenue'] = leads_df['Company Revenue'].astype(int)
```

```python
leads_df.rename(columns={' Empoyee Head Count in millions': 'Employee Head Count'}, inplace=True)
leads_df.rename(columns={'Company Revenue': 'Company Revenue (in millions)'}, inplace=True)
leads_df.drop_duplicates(subset='Company Name', keep='first', inplace=True)
leads_df.reset_index(drop=True, inplace=True)
```
